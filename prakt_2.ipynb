```python
import pandas as pd
import scipy
import numpy as np 
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('diabetes.csv')
print(df.head())

df.info()

df.isnull().sum()

df.describe()

fig, axs = plt.subplots(9,1,dpi=95, figsize=(7,17))

i = 0
for col in df.columns:
    axs[i].boxplot(df[col], vert=False)
    axs[i].set_ylabel(col)
    i+=1
plt.show()

q1, q3 = np.percentile(df['Insulin'],[25,75])
iqr = q3-q1
lower_bound = q1 - (1.5*iqr)
upper_bound = q3 + (1.5*iqr)
clean_data = df[(df['Insulin']>=lower_bound) & (df['Insulin']<=upper_bound)]

q1, q3 = np.percentile(df['Pregnancies'],[25,75])
iqr = q3-q1
lower_bound = q1 - (1.5*iqr)
upper_bound = q3 + (1.5*iqr)
clean_data = df[(df['Pregnancies']>=lower_bound) & (df['Pregnancies']<=upper_bound)]

q1, q3 = np.percentile(df['Age'],[25,75])
iqr = q3-q1
lower_bound = q1 - (1.5*iqr)
upper_bound = q3 + (1.5*iqr)
clean_data = df[(df['Age']>=lower_bound) & (df['Age']<=upper_bound)]

q1, q3 = np.percentile(df['Glucose'],[25,75])
iqr = q3-q1
lower_bound = q1 - (1.5*iqr)
upper_bound = q3 + (1.5*iqr)
clean_data = df[(df['Glucose']>=lower_bound) & (df['Glucose']<=upper_bound)]

q1, q3 = np.percentile(df['BloodPressure'],[25,75])
iqr = q3-q1
lower_bound = q1 - (1.5*iqr)
upper_bound = q3 + (1.5*iqr)
clean_data = df[(df['BloodPressure']>=lower_bound) & (df['BloodPressure']<=upper_bound)]

q1, q3 = np.percentile(df['BMI'],[25,75])
iqr = q3-q1
lower_bound = q1 - (1.5*iqr)
upper_bound = q3 + (1.5*iqr)
clean_data = df[(df['BMI']>=lower_bound) & (df['BMI']<=upper_bound)]

q1, q3 = np.percentile(df['DiabetesPedigreeFunction'],[25,75])
iqr = q3-q1
lower_bound = q1 - (1.5*iqr)
upper_bound = q3 + (1.5*iqr)
clean_data = df[(df['DiabetesPedigreeFunction']>=lower_bound) & (df['DiabetesPedigreeFunction']<=upper_bound)]



corr = df.corr()
plt.figure(dpi=130)
sns.heatmap(df.corr(), annot=True, fmt= '.2f')
plt.show()

plt.pie(df.Outcome.value_counts(),
labels= ['Diabetes', 'Not Diabetes'],
autopct='%.f', shadow=True)
plt.title('Outcome Proportionality')
plt.show()


```
